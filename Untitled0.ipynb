```python
df.loc[df["bill_length_mm"] > 42, ["break_length"]] = "max_break_length"
df.loc[(df["bill_length_mm"] > 35) & (df["bill_length_mm"] < 42), ["break_length"]] = "sr_break_length"
df.loc[df["bill_length_mm"] < 35, ["break_length"]] = "min_break_length"
```

```python
df.loc[df['bill_length_mm'] >= 42, 'height_group'] = 'high'
df.loc[(df['bill_length_mm'] < 42) & (df['bill_length_mm'] > 35), 'height_group'] = 'middle'
df.loc[df['bill_length_mm'] < 35 , 'height_group'] = 'low'
df.head(20)
sns.displot(data = df, x = 'flipper_length_mm', hue = 'height_group')

```
